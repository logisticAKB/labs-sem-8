```python
import json
import shutil, os 
from io import StringIO

import numpy as np
import pandas as pd
from apyori import apriori, dump_as_json
```

```python
transactions = []
with open('retail.dat', 'r') as file:
    for i, line in enumerate(file):
        transactions.append(line.strip().split())
```

```python
def display_result(result):
    output = []
    for RelationRecord in result:
        o = StringIO()
        dump_as_json(RelationRecord, o)
        output.append(json.loads(o.getvalue()))

    for i_output in range(len(output)):
        output[i_output]['ordered_statistics'] = sorted(output[i_output]['ordered_statistics'], key=lambda x: x['confidence'], reverse=True)

    data_df = pd.DataFrame(output)
    data_df = data_df.sort_values(by='support', ascending=False)

    pd.set_option('display.max_colwidth', None)

    from IPython.display import display, HTML
    display(HTML(data_df.to_html()))
```

## Min support - 0.3%, min confidence - 20%

```python
result = list(apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 4, min_length = 2))
display_result(result)
```
